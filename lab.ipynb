```python
#Flights 1
import pandas as pd
flights = pd.read_csv("flights.csv")

cleaned_flights = flights.dropna(subset=["dep_delay", "arr_delay"]) 
cleaned_flights = cleaned_flights[(cleaned_flights["dep_delay"] != "NA") & (cleaned_flights["arr_delay"] != "NA")]
cleaned_flights = cleaned_flights.reset_index(drop=True)


cleaned_flights.to_csv("flights_cleaned.csv", index=False)


```

```python
#check for NA values in dep_delay and arr_delay
import pandas as pd

flights = pd.read_csv("flights_cleaned.csv")

has_na = ((flights["dep_delay"] == "NA") | (flights["arr_delay"] == "NA")).any()

has_missing = (flights["dep_delay"].isna() | flights["arr_delay"].isna()).any()

if has_na or has_missing:
    print("There are NA or missing values in dep_delay or arr_delay columns.")
else:
    print("There are no NA or missing values in dep_delay or arr_delay columns.")
```

```python
#Flights 2
import pandas as pd
cleaned_flights = pd.read_csv("flights_cleaned.csv")

cleaned_flights["dep_delay"] = pd.to_numeric(cleaned_flights["dep_delay"], errors="coerce")
cleaned_flights["arr_delay"] = pd.to_numeric(cleaned_flights["arr_delay"], errors="coerce")

mean_delays = cleaned_flights.groupby(["carrier", "flight"], as_index=False)[["dep_delay", "arr_delay"]].mean()

print(mean_delays.head())

mean_delays.to_csv("flight_mean.csv", index=False)
```

```python
#Wine 1

import pandas as pd
import matplotlib.pyplot as plt

wine = pd.read_csv("Wine.csv")

plt.figure(figsize=(10,5))
plt.hist(wine['total sulfur dioxide'], bins=20, color='lightblue', edgecolor='black')
plt.title("Histogram: Total Sulfur Dioxide")
plt.xlabel("Total Sulfur Dioxide")
plt.ylabel("Frequency")
plt.show()

print("This histogram show the distribution of total sulfur dioxide. As we can see, the histogram is skewed to the right. We can observe how most wines contain low amounts of sulfur dioxide. However, there are few with high amounts of sulfur dioxide")
```

```python
#Wine 2
import pandas as pd
import matplotlib.pyplot as plt

wine = pd.read_csv("Wine.csv")
plt.figure(figsize=(10,5))
plt.scatter(wine['residual sugar'], wine['quality'], alpha=0.5, color='green')
plt.title("Residual Sugar vs Wine Quality")
plt.xlabel("Residual Sugar")
plt.ylabel("Quality")
plt.show()
print("There is no clear linear relationship between residual sugar and quality. Hence, sweetness does not determine wine quality.")
```

```python
#Wine 3
import pandas as pd
import matplotlib.pyplot as plt

wine = pd.read_csv("Wine.csv")

plt.figure(figsize=(8,5))
plt.hexbin(wine['residual sugar'], wine['alcohol'], gridsize=15, cmap='cividis', edgecolor='black')
plt.title("Hexbin Plot: Residual Sugar vs Alcohol")
plt.xlabel("Residual Sugar")
plt.ylabel("Alcohol")
plt.colorbar(label='Count')
plt.show()
print("Most wines cluster around low sugar and moderate alcohol. There appears to be slight negative relationship between the amount of sugar and alcohol. The more alcohol the less sugar there is. This is because of the fermentation process; sugar conversion into alcohol.")
```
